```python
import pandas as pd
import cv2
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import resample

# Load the CSV file
csv_path = "/content/drive/MyDrive/OCT-AND-EYE-FUNDUS-DATASET-main/fundusl.csv"
df = pd.read_csv(csv_path)

# Initialize empty lists for image data and class labels
images = []
labels = []

# Define the target image size
target_size = (256, 256)  # Adjust as per your requirement

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Extract the image name and class label
    image_name = row['Name']
    class_label = row['class']

    # Load the image using OpenCV or any other image processing library
    image_path = "/content/drive/MyDrive/OCT-AND-EYE-FUNDUS-DATASET/EYE FUNDUS/" + image_name + ".jpg"
    image = cv2.imread(image_path)

    # Resize the image to the target size
    resized_image = cv2.resize(image, target_size)

    # Perform preprocessing on the image (e.g., normalize, feature extraction)
    # ...

    # Append the processed image and class label to the lists
    images.append(resized_image)
    labels.append(class_label)

# Convert the image and label lists to numpy arrays
X = np.array(images)
y = np.array(labels)

# Oversample the minority classes using Random Oversampling
X_balanced = []
y_balanced = []
class_counts = {}

# Iterate over each unique class label
for class_label in np.unique(y):
    # Get the indices of samples belonging to the current class
    indices = np.where(y == class_label)[0]

    # Resample the minority class samples to match the majority class
   # Determine the maximum number of samples to be resampled
    max_samples = np.max(list(class_counts.values()))

# Resample the minority class samples to match the majority class
    oversampled_indices = resample(indices, replace=True, n_samples=max_samples, random_state=42)


    # Append the oversampled samples to the balanced data
    X_balanced.extend(X[oversampled_indices])
    y_balanced.extend(y[oversampled_indices])

    # Update the class count dictionary
    class_counts[class_label] = len(oversampled_indices)

# Convert the balanced data to numpy arrays
X_balanced = np.array(X_balanced)
y_balanced = np.array(y_balanced)

# Create a new DataFrame with the balanced data
balanced_df = pd.DataFrame({'image_name': [f'image_{i+1}' for i in range(len(X_balanced))], 'class': y_balanced})

# Save the balanced dataset to a new CSV file
new_csv_path = "/content/drive/MyDrive/OCT-AND-EYE-FUNDUS-DATASET-main/fundusoversampled.csv"
balanced_df.to_csv(new_csv_path, index=False)

# Plot the count and class distribution after balancing
class_counts = balanced_df['class'].value_counts()
class_labels = class_counts.index

plt.bar(class_labels, class_counts)
plt.xlabel('Class')
plt
```
